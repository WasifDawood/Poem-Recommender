```python
import os  #accessing my operating system and navigation 
import pandas as pd #for data storage and analysis
import json #for reading and writing json files
from tqdm import tqdm #Progress bar for better visual representation of process completion 


def body_to_text(body): 
  current_stanza_id = 0
  body_text = ""
    
  for line in body:
      
    if current_stanza_id < line['stanza_id']:
      current_stanza_id += 1
      body_text = body_text + "\n"
    body_text = body_text + "\n" + line['text']

  return body_text

df_rows = []

for file in tqdm(os.listdir("en")): #looping through each file in the directory "en" 

  #print(file) #prints the file name

  with open("en/" + file) as j: #for proper handling of files
  
    data = json.load(j) #parses the JSON file into a python dictionary

    duplicate = data.get("duplicate", False)

    if duplicate: 
      continue 

    poem_id = data.get("id")
    title = data.get("title")
    year_created = data.get("year_created")
    neighbors = data.get("neighbors", [])
    

    author = data.get("author", {})
    
    if author is None :
      author_name = author_country = author_born = author_died = ""
      #TODO if list take first author
    
    elif isinstance(author, list):  
      if len(author) > 0:
        first_author = author[0]  
        author_country = first_author.get("country", "")
        author_born = first_author.get("born", "")
        author_died = first_author.get("died", "")
        
    else : 
      author_name = author.get("name")
      author_country = author.get("country")
      author_born = author.get("born")
      author_died = author.get("died")


    source = data.get("source", {})
    source_place = source.get("place")
    
    body = data.get("body", [])

    body_text = body_to_text(body)

    df_rows.append(

      {
        "poem_id" : poem_id,
        "title" : title,
        "year_created" : year_created,
        "neighbors" : neighbors,
        "duplicate" : duplicate,
        "author_name" : author_name,
        "author_country" : author_country,
        "source_place" : source_place,
        "body_text" : body_text,
        "author_born" : author_born,
        "author_died" : author_died
      }
    )
    


  
df = pd.DataFrame(df_rows)
print(df)
df.to_csv("PoemData.csv", index = False)


  

```
